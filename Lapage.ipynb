## On vérife s'il y a des valeurs impossibles dans le df_products

```python
# Identification des valeurs anormales  
for i in range (len(df_products['price'])):
    if df_products['price'][i] < 0:
        print(df_products['id_prod'][i])
```

## On note qu'il y a des valeurs aberrantes T_0 dans "id_prod" et dans "prices" on a une valeur négative dans le df_products 

```python
# Dans tableau de 'products' Sélectionner la colonne qui a les valeurs aberrantes 
df_products.set_index('id_prod', inplace=True)
df_products.index
```

```python
# Observer toutes les informations du valeur T_0 
df_products.loc['T_0']
```

```python
# Remplacer la valeur aberrante 'T_0' à 'NaN' (forme de le valeur manquante) dans dataframe "products"
df_products.reset_index(inplace=True)
import numpy
df_products.replace('T_0',numpy.nan,inplace= True)
print(df_products.isnull().sum())
# "T_0" a été remplacé par le valeur manquante
```

```python
# Supprimer la valeur aberrante dans "products"
df_products = df_products.dropna()
print(df_products.isnull().sum())
# Le valeur manquante a été supprimé
```

# **La fiche produit manque pour le produit id = 0_2245 vendu 103 fois, que pensez vous?**

```python
# Vérifier si le produit 0_2245 est dans df_trans
df_trans[df_trans['id_prod']=='0_2245']
# On constate que le produit id = 0_2245 s'affiche dans df_trans mais ne s'affiche pas dans df_products
# ce qui veut dire que le produit id = 0_2245 est une valeur manquante du df_products.
```

## L'évolution du chiffre d'affaire par temps

```python
# Croiser la table d'évolution de CA par chaque mois
df_evolutionCA = df.groupby('mois_annee').sum()['price'].reset_index()
df_evolutionCA
```

```python
# Créer le graphique ligne pour visualiser l'évolution de CA par temps
%matplotlib inline
plt.figure(figsize=(15,6))
plt.plot(df_evolutionCA['mois_annee'], df_evolutionCA['price'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.title("l'évolution de CA par temps")
plt.show()
```

# La courbe de Lorenz

```python
# La table des CAs par age
CAage = df.groupby(["age"])['price'].sum().reset_index()
CAage
```

```python
# Transformer les données sous forme d'array afin de créer la courbe de Lorenz 
import numpy as np
X = np.array(CAage['price'])
X
```

```python
# Création le courbe de lorenz CA
t_number = CAage['price'].values
lorenz = np.cumsum(np.sort(X)) / X.sum()
lorenz = np.append([0], lorenz)
plt.plot(np.linspace(0,1,len(lorenz)), lorenz, drawstyle='steps-post')
plt.plot(np.linspace(0,1,len(lorenz)), np.linspace(0,1,len(lorenz)), dashes=[2,4])
plt.ylabel('part cumulée du CA de produits')
plt.xlabel('part cumulée des âges des clients')
plt.title("Chiffre d'affaires")
aire_ss_courbe = lorenz[:-1].sum() / len(lorenz)
S = 0.5 - aire_ss_courbe
gini = 2*S
plt.savefig('courbe de lorenz ca.png')
print(f'indice de gini: {gini:.2f}')
```
